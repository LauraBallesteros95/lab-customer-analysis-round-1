```python
#Show the DataFrame's shape.
import pandas as pd

# read the CVS and asign the result to the variables df1, df2 y df3
df1 = pd.read_csv('C:/Users/laura/Desktop/ironhack/lab-customer-analysis-round-1/files_for_lab/csv_files/file1.csv')
df2 = pd.read_csv('C:/Users/laura/Desktop/ironhack/lab-customer-analysis-round-1/files_for_lab/csv_files/file2.csv')
df3 = pd.read_csv('C:/Users/laura/Desktop/ironhack/lab-customer-analysis-round-1/files_for_lab/csv_files/file3.csv')

# print the result of df3
print(df3.shape)
```

```python
#Standardize header names.
df1.columns = df1.columns.str.lower().str.replace(" ", "_")
df2.columns = df2.columns.str.lower().str.replace(" ", "_")
df3.columns = df3.columns.str.lower().str.replace(" ", "_")

df3 = df3.rename(columns={'state':'st'})
```

```python
df1_order = ['customer', 'gender', 'education','vehicle_class', 'st', 'income','policy_type', 'customer_lifetime_value', 'monthly_premium_auto', 'number_of_open_complaints', 'total_claim_amount']
ord_df1 = df1[df1_order]
df2_order = ['customer', 'gender', 'education','vehicle_class', 'st', 'income','policy_type', 'customer_lifetime_value', 'monthly_premium_auto', 'number_of_open_complaints', 'total_claim_amount']
ord_df2 = df2[df2_order]
df3_order = ['customer', 'gender', 'education','vehicle_class', 'st', 'income','policy_type', 'customer_lifetime_value', 'monthly_premium_auto', 'number_of_open_complaints', 'total_claim_amount']
ord_df3 = df3[df3_order]
```

```python
#Concatenate the three dataframes
data = pd.concat([ord_df1, ord_df2, ord_df3]).reset_index(drop=True)
data
```

Understand the meaning of all columns

-customer: Categorical: the ID of every customer.
-st: Categorical: the state of each customer
-gender: Categorical: represents the gender of each customer
-education: Categorical: the education level of every customer
-customer_lifetime_value : Numerical:  the value of the customer over the lifetime
-income : Numerical :the income of every customer
-monthly_premium_auto :Numerical: monthly payment for premium users
-number_of_open_complaints:Numerical :the number of complaints that each customer has open
-policy_type : Categorical : the type of policy thaht customer has contracted.
-vehicle_class :Categorical :which type of vehicle the customer has
-total_claim_amount: Numerical :the total amount of money for all the complaints open for each custom

```python
#Delet the columns education and number of open complaints
data = data.drop('education', axis=1)
data = data.drop('number_of_open_complaints', axis=1)
```

```python
#Correct the values of the column customer lifetime value

data['customer_lifetime_value'] = data['customer_lifetime_value'].str.rstrip('%').astype(float) / 100
data
```

```python
#Check  duplicates and delet 
data.drop_duplicates()
```

```python
#Filtering and printing the customers that have a value of 0 or less in income.

income_filter = data[(data['income'] == 0)]
print(income_filter)
```
